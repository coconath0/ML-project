**Project: Water Quality**

First, we'll obtain the dataset by importing openML and saving it on /data/ folder.

This following code will only be run once to obtain the dataset!

```python
# Cell 1: Install OpenML and import necessary libraries
#%pip install openml pandas -q 
# Uncomment the line above if you get an error that 'openml' is not installed

import openml
import pandas as pd
import os

# 1. Define the dataset ID and file path
DATASET_ID = 46085 
OUTPUT_FOLDER = 'data'
OUTPUT_FILE = os.path.join(OUTPUT_FOLDER, 'water-quality-46085.csv')

# 2. Fetch the dataset using its ID
print(f"Fetching OpenML Dataset ID: {DATASET_ID}...")
dataset = openml.datasets.get_dataset(DATASET_ID)

# 3. Get the data as a Pandas DataFrame
X, y, categorical_indicator, attribute_names = dataset.get_data(
    dataset_format="dataframe",
    target=dataset.default_target_attribute  # Ignore target for now, get all columns
)

# Combine features (X) and the target (y) if a target was defined, or just use X
# Since this dataset is complex, we just use the X DataFrame which contains all features
df_raw = X.copy()
print(f"Data fetched successfully. Shape: {df_raw.shape}")

# 4. Save the DataFrame to your local data folder
os.makedirs(OUTPUT_FOLDER, exist_ok=True) # Creates the 'data' folder if it doesn't exist
df_raw.to_csv(OUTPUT_FILE, index=False)
print(f"\nData saved locally to: {OUTPUT_FILE}")
```
